```python
def get_status(data, ctg):
    df_ctg = data[data['Группа2'] == ctg]
    mean_marga_gds = round(df_ctg['Маржа'].sum() / df_ctg['КоличествоПроданоКлиенту'].sum(), 2)
    mean_price_gds = round(df_ctg['СуммаСтроки'].sum() / df_ctg['КоличествоПроданоКлиенту'].sum(), 2)
    max_gain_gds = round(mean_marga_gds / mean_price_gds * 100, 2)
    
    num_rcpt = len(df_ctg['НомерЗаказаНаСайте'].unique()) 
    
    mean_marga_rcpt = round(df_ctg['Маржа'].sum() / num_rcpt, 2)
    mean_price_rcpt = round(df_ctg['СуммаСтроки'].sum() / num_rcpt, 2)
    max_gain_rcpt = round(mean_marga_rcpt / mean_price_rcpt * 100, 2)
    
    res = (
            mean_marga_gds, mean_price_gds, max_gain_gds,
            mean_marga_rcpt, mean_price_rcpt, max_gain_rcpt
    )
    return res

def get_support(data, item_left, item_right=None):
    if item_right is None:
        return data[(data[item_left] != 0.)].shape[0] / data.shape[0]
    return data[(data[item_left] != 0.) & (data[item_right] != 0.)].shape[0] / data.shape[0]

def build_mba(data, receipts_data):
    results = pd.DataFrame(columns=['item_left', 'item_right', 'support', 'confidence', 'lift', 
                                    'num_target', 
#                                     'impact_0', 'impact_5', 'impact_10', 'impact_15', 'impact_20',
#                                     'mean_marga_gds', 'mean_price_gds', 'max_gain_gds', 
                                    'mean_marga_rcpt', 'mean_price_rcpt', 'max_gain_rcpt'])
    stats = dict()
    for item in tqdm(COLUMNS):
        mean_marga_gds, mean_price_gds, max_gain_gds, \
        mean_marga_rcpt, mean_price_rcpt, max_gain_rcpt = get_status(receipts_data, item)
        stats[item] = (mean_marga_gds, mean_price_gds, max_gain_gds,
                      mean_marga_rcpt, mean_price_rcpt, max_gain_rcpt)
    
    for item_left in tqdm(COLUMNS):
        for item_right in tqdm(COLUMNS, leave=False):
            if item_left == item_right:
                continue
            sup = get_support(data, item_left, item_right)
            try:
                conf = sup / get_support(data, item_left)
            except:
                conf = np.nan
            try:
                lift = conf / get_support(data, item_right)
            except:
                lift = np.nan

            num_target = data[data[item_left] != 0.].shape[0] - data[(data[item_left] != 0.) 
                                                                     & (data[item_right] != 0.)].shape[0]
#             impact_0 = num_target * stats[item_right][3]
#             impact_5 = num_target * stats[item_right][3] * (stats[item_right][5] - 5) / (stats[item_right][5])
#             impact_10 = num_target * stats[item_right][3] * (stats[item_right][5] - 10) / (stats[item_right][5])
#             impact_15 = num_target * stats[item_right][3] * (stats[item_right][5] - 15) / (stats[item_right][5])
#             impact_20 = num_target * stats[item_right][3] * (stats[item_right][5] - 20) / (stats[item_right][5])
            # conv = (1 - sup) / (1 - conf)
            results = results.append({'item_left' : item_left,
                            'item_right' : item_right,
                            'support' : sup,
                            'confidence' : conf,
                            'lift' : lift,
                            'num_target' : num_target,
#                             'impact_0' : impact_0,
#                             'impact_5' : impact_5,
#                             'impact_10' : impact_10,
#                             'impact_15' : impact_15,
#                             'impact_20' : impact_20,
#                             'mean_marga_gds' : stats[item_right][0],
#                             'mean_price_gds' : stats[item_right][1],
#                             'max_gain_gds' : stats[item_right][2],
                            'mean_marga_rcpt' : stats[item_right][3],
                            'mean_price_rcpt' : stats[item_right][4],
                            'max_gain_rcpt' : stats[item_right][5]}, ignore_index=True)
    return results
```

```python
def get_sums(data):
    
    #### TO DO: lift > 1 смотрим по ним кто что покупал и ебашим им отправку
    
    results = pd.DataFrame(columns=['item', 'support', 'confidence', 'lift', 
                                    #'mean_marga_gds', 'mean_price_gds', 'max_gain_gds',
                                    'mean_marga_rcpt', 'mean_price_rcpt', 'max_gain_rcpt',
                                    'mean_target'])
    for item in tqdm(data['item_right'].unique()):
        tmp = data[data['item_right'] == item]
        
        results = results.append({'item' : item,
                            'support' : tmp['support'].sum(),
                            'confidence' : tmp['confidence'].sum(),
                            'lift' :tmp['lift'].sum(),
                            'mean_marga_rcpt' : tmp['mean_marga_rcpt'].mean(),
                            'mean_price_rcpt' : tmp['mean_price_rcpt'].mean(),
                            'max_gain_rcpt' : tmp['max_gain_rcpt'].mean(),
                            'mean_target' : tmp['num_target'].mean()}, ignore_index=True)
    return results
```
