### Функция подготовки данных ###
После вышеизложенного исследования данных можно сформулировать что должна делать первая функция по отчистке и подготовке данных:

* Удалить все строчки где 'Номенклатура' = 'Доставка'
* Удалить все строчки где пропущены значения в колонках 'ЦенаЗакупки', 'Цена' или 'Количество'
* Преобразовать колонки в правильный тип
* Пересчитать правильно маржу и сумму строки
* Удалить дубликаты если они есть
* Удалить фрод, то есть чеки в которых присутсвуют товары с большой отрицательной маржой
* Удалить колонки 'СуммаЗаказаНаСайте', 'СуммаДокумента' чтобы избежать неправильной трактовки в дальнейшем

```python
def object_columns_to_float_columns(data, column_names):
    for column_name in column_names:
        data[column_name] = data[column_name].apply(lambda x : x.replace(',', '.'))
        data[column_name] = data[column_name].apply(lambda x : x.replace(' ', ''))
        data[column_name] = pd.to_numeric(data[column_name])
    return data
```

```python
def prepare_data(data):
    ### TO DO: check month and years
    ### TO DO: reset_indices
    
    current_number_of_raws = data.shape[0]
    data = data[data['Номенклатура'] != 'Доставка']
    print('Количество удаленных строк, которые описывают доставку: ', current_number_of_raws - data.shape[0])
    current_number_of_raws = data.shape[0]
    data = data.dropna(subset=['Количество', 'Цена', 'ЦенаЗакупки'])
    print('Количество удаленных строк, где пропущены ключевые значения: ', current_number_of_raws - data.shape[0])
    
    data = object_columns_to_float_columns(data, ['СуммаДокумента', 'СуммаЗаказаНаСайте', 'СуммаСтроки', 'ЦенаЗакупки', 'Цена'])
    
    data['Дата'] = pd.to_datetime(data['Дата'], format='%d.%m.%Y %H:%M')
    data['ДатаДоставки'] = pd.to_datetime(data['ДатаДоставки'], format='%d.%m.%Y %H:%M')
    data['ДатаЗаказаНаСайте'] = pd.to_datetime(data['ДатаЗаказаНаСайте'], format='%d.%m.%Y %H:%M')
    
    data['ДеньДатыЗаказа'] = data['Дата'].dt.dayofweek
    data['ЧасДатыЗаказа'] = data['Дата'].dt.hour
    
    data['СуммаСтроки'] = data['Цена'] * data['Количество']
    data['Маржа'] = (data['Цена'] - data['ЦенаЗакупки']) * data['Количество']
    data
    
    current_number_of_raws = data.shape[0]
    data = data.drop_duplicates(subset=data.columns, keep='last')
    print('Количество удаленных строк, которые были дубликатами: ', current_number_of_raws - data.shape[0])
    
    current_number_of_raws = data.shape[0]
    fraud_ids = set(data[data['Маржа'] <= -300.0]['НомерЗаказаНаСайте'].unique())
    data = data[~data['НомерЗаказаНаСайте'].isin(fraud_ids)]
    print('Количество удаленных строк, которые являются фродом: ', current_number_of_raws - data.shape[0])
    print('Количество удаленных чеков, которые являются фродом: ', len(fraud_ids))
    
    data = data.drop(columns=['СуммаЗаказаНаСайте', 'СуммаДокумента'])
    
    print('Итоговое количество строк: ', data.shape[0])
    return data
```

### Выделяем выкупленные товары

Для начала посмотрим какие значения принимает НовыйСтатус:

```python
print(data['НовыйСтатус'].unique())
```

<p>Исходя из этих данных нетрудно понять, что заказ на пути от корзины сайта до получателя проходит достаточно длинный путь. Будем считать, что выкупленные товары это только те у которых 'НовыйСтатус' равно 'Доставлен'.<br><br>Теперь посмотрим какие значения в таком случае принимает колонка 'Отменено'.</p>

```python
data[data['НовыйСтатус'] == 'Доставлен'].groupby('Отменено').size()
```

<p>Как можно заметить, даже при 'НовыйСтатус' = 'Доставлен' у нас 'Отменено' может принимать значение 'Нет'. Мы считаем, что колонка 'Отменено' первична, следовательно, заказы которые доставлены, но отменены мы учитывать не будем.</p>
<p>Таким образом мы можем сделать вывод какие товары являются выкупленными:</p>

Теперь посмотрим на 'Количество' и 'КоличествоПроданоКлиенту', узнаем существуют ли варианты где они не равны

```python
data[data['Количество'] != data['КоличествоПроданоКлиенту']].head()
```

Заметим, что если 'Количество' не равно 'КоличествоПроданоКлиенту', то 'СуммаСтроки' и 'Маржа' пересчитываются по 'Количество' следовательно, чтобы исправить корректность данных, анализируя выкупленные товары, нам необходимо пересчитать их по 'КоличествоПроданоКлиенту'. Давайте сразу удалим те строки в которых 'КоличествоПроданоКлиенту' равно 0.

Таким образом чтобы выделить выкупленные товары нам нужно:
* Оставить только те строки, где 'НовыйСтатус' == 'Доставлен', а 'Отменено' == 'Нет'
* Удалить все строчки, где пропущено значение в столбце 'КоличествоПроданоКлиенту'
* Удалить все строчки, где 'КоличествоПроданоКлиенту' <= 0
* Пересчитать основные показатели через КоличествоПроданоКлиенту, а не через Количество

```python
def get_purchased_goods(data):
    current_number_of_raws = data.shape[0]
    data = data[data['НовыйСтатус'] == 'Доставлен'][data['Отменено'] == 'Нет']
    print('Количество удаленных строк, где товар не доставлен или отменен: ', current_number_of_raws - data.shape[0])
    
    current_number_of_raws = data.shape[0]
    data = data.dropna(subset=['КоличествоПроданоКлиенту'])
    print('Количество удаленных строк, где пропущено значение в колонке \'КоличествоПроданоКлиенту\':', current_number_of_raws - data.shape[0])
    
    current_number_of_raws = data.shape[0]
    data = data[data['КоличествоПроданоКлиенту'] > 0]
    print('Количество удаленных строк, где \'КоличествоПроданоКлиенту\' <= 0: ', current_number_of_raws - data.shape[0])
    
    data['СуммаСтроки'] = data['Цена'] * data['КоличествоПроданоКлиенту']
    data['Маржа'] = (data['Цена'] - data['ЦенаЗакупки']) * data['КоличествоПроданоКлиенту']
    
    print('Итоговое количество строк: ', data.shape[0])
    return data

purchased_data = get_purchased_goods(data)
```

Функция ниже получает на вход срез данных и считает основные показатели такие как:

* Выручка
* Кол-во уникальных чеков
* Средний чек
* Средняя цена
* Средняя маржа
* Средний % маржи
* Кол-во товаров
* Кол-во уникальных клиентов
* Среднее количество товаров в чеке

```python
def get_basic_indicators(data, is_purchased):
    result = pd.DataFrame(np.nan, index=[0], columns=['Выручка', 'Кол-во уникальных чеков', 
                                   'Средний чек', 'Средняя цена', 'Средняя маржа', 
                                   'Средний % маржи', 'Кол-во товаров', 'Кол-во уникальных чеков',
                                   'Среднее количество товаров в чеке'])
    if is_purchased:
        number = 'КоличествоПроданоКлиенту'
    else:
        number = 'Количество'
    
    result['Выручка'] = data['СуммаСтроки'].sum()
    result['Кол-во уникальных чеков'] = len(data['НомерЗаказаНаСайте'].unique())
    
    tmp_table = data.groupby(['НомерЗаказаНаСайте'])['СуммаСтроки'].sum()
    result['Средний чек'] = tmp_table.mean()
    
    
    result['Средняя цена'] = (data['Цена'] * data[number]).sum() / data[number].sum()
    
    result['Средняя маржа'] = data['Маржа'].sum() / data[number].sum()
    
    result['Средний % маржи'] = (((data['Цена'] - data['ЦенаЗакупки']) / data['ЦенаЗакупки'])
                                 * data[number]).sum() / data[number].sum()
    
    tmp_table = data.groupby(['НомерЗаказаНаСайте'])[number].sum()
    
    result['Среднее количество товаров в чеке'] = tmp_table.mean()
    result['Кол-во товаров'] = tmp_table.sum()
    result['Кол-во уникальных чеков'] = tmp_table.shape[0]
    return result
```

### Основные показатели по месяцам ###

```python
def get_basic_indicators_by_month(data, is_purchased=False):
    result = pd.DataFrame(columns=['Выручка', 'Кол-во уникальных чеков', 
                                   'Средний чек', 'Средняя цена', 'Средняя маржа', 
                                   'Средний % маржи', 'Кол-во товаров', 'Кол-во уникальных чеков',
                                   'Среднее количество товаров в чеке'])
    for current_month in range(7, 11):
        current_data = data[data['МесяцДатыЗаказа'] == current_month]
        result = result.append(get_basic_indicators(current_data, is_purchased), ignore_index=True)
    return result
```
