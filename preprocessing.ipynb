```python
order_data1[order_data1['НовыйСтатус'] == 'Отменен'].groupby('Отменено').count()['Дата']

### Среди заказов, у которых 'НовыйСтатус' == 'Отменен', есть те,
### у которых 'Отменено' == 'Нет' причем их довольно много.
```

```python
order_data1[order_data1['Отменено'] == 'Да'].groupby('НовыйСтатус').count()['Дата']

### Смотрим на распределение заказов по статусам среди отмененных.
### Не у всех статус 'Отменен'.
```

```python
order_data1[order_data1['Отменено'] == 'Нет'].groupby('НовыйСтатус').count()['Дата']

### Смотрим на распределение заказов по статусам среди неотмененных.
### Больше всего со статусом 'Доставлен', но есть и много
### со статусами 'Возврат из ПВЗ', а также 'Отменен'.
```
