```python
INDICATORS = [
    'КоличествоЧеков', 'КоличествоЧековВсе', 'СреднийЧек',
    'СреднееЧислоТоваровЧека', 'СреднееЧислоТоваровЧекаВсе',
    'Выручка', 'ВыручкаВся', 'КоличествоТоваров',
    'КоличествоТоваровВсе', 'СредняяМаржа', 'СредняяМаржаВся',
    'ДоляВыкупленных', 'ПроцентОтмен', 
    'ДатаПоследнегоЗаказаВЫК', 'ДатаПоследнегоЗаказа'
]
TYPE = list(data['Тип'].unique())
GROUP2 = list(data['Группа2'].unique())
GROUP3 = list(data['Группа3'].unique())
REGION = list(data['Регион'].unique())
PAYMENT_FORM = list(data['ФормаОплаты'].unique())
DAY = list(data['ДеньДатыЗаказа'].unique())
TIME = list(data['ЧасДатыЗаказа'].unique())
SHIPPING_METHOD = list(data['МетодДоставки'].unique())

COLUMNS = (INDICATORS + TYPE + GROUP2 + GROUP3 + REGION +
           PAYMENT_FORM + DAY + TIME + SHIPPING_METHOD)
```

```python
def client_basket(data, type_name, types):
    result = pd.DataFrame(np.nan, index=[0], columns=types)
    
    groups = data.groupby(type_name)
    groups_count = groups['КоличествоПроданоКлиенту'].sum()
    total_count = data['КоличествоПроданоКлиенту'].sum()

    for t in types:
        try:
            count = groups_count[t]
        except KeyError:
            count = 0
            
        result[t] = count / total_count
        
    return result

def client_info(data):
    result = pd.DataFrame(np.nan, index=[0], columns=COLUMNS)

    purchased_data = data[(data['Выкуплен'] == 'Да')]
    all_receipts = data.groupby(['НомерЗаказаНаСайте'])
    purchased_receipts = purchased_data.groupby(['НомерЗаказаНаСайте'])

    result['КоличествоЧеков'] = purchased_receipts.count().shape[0]
    result['КоличествоЧековВсе'] = all_receipts.count().shape[0]

    result['СреднийЧек'] = purchased_receipts['СуммаСтроки'].sum().mean()

    result['СреднееЧислоТоваровЧека'] = purchased_receipts['КоличествоПроданоКлиенту'].sum().mean()
    result['СреднееЧислоТоваровЧекаВсе'] = all_receipts['Количество'].sum().mean()

    result['Выручка'] = purchased_data['СуммаСтроки'].sum()
    result['ВыручкаВся'] = data['СуммаСтроки'].sum()

    result['КоличествоТоваров'] = purchased_data['КоличествоПроданоКлиенту'].sum()
    result['КоличествоТоваровВсе'] = data['Количество'].sum()

    result['СредняяМаржа'] = purchased_data['Маржа'].sum() / result['КоличествоТоваров']
    result['СредняяМаржаВся'] = data['Маржа'].sum() / result['КоличествоТоваровВсе']

    result['ДоляВыкупленных'] = purchased_data.shape[0] / data.shape[0]

    result['ПроцентОтмен'] = data[(data['НовыйСтатус'] == 'Отменен') |
                                 (data['Отменено'] == 'Да')].shape[0] / data.shape[0]

    result['ДатаПоследнегоЗаказа'] = np.min(data['ДнейСЗаказа'].values)

    if purchased_data['ДнейСЗаказа'].shape[0] > 0:
        result['ДатаПоследнегоЗаказаВЫК'] = np.min(purchased_data['ДнейСЗаказа'].values)
    else:
        result['ДатаПоследнегоЗаказаВЫК'] = np.min(data['ДнейСЗаказа'].values)

    result.loc[:, TYPE] = client_basket(purchased_data, 'Тип', TYPE)
    result.loc[:, GROUP2] = client_basket(purchased_data, 'Группа2', GROUP2)
    result.loc[:, GROUP3] = client_basket(purchased_data, 'Группа3', GROUP3)
    result.loc[:, REGION] = client_basket(purchased_data, 'Регион', REGION)
    result.loc[:, PAYMENT_FORM] = client_basket(purchased_data, 'ФормаОплаты', PAYMENT_FORM)
    result.loc[:, DAY] = client_basket(purchased_data, 'ДеньДатыЗаказа', DAY)
    result.loc[:, TIME] = client_basket(purchased_data, 'ЧасДатыЗаказа', TIME)
    result.loc[:, SHIPPING_METHOD] = client_basket(purchased_data, 'МетодДоставки', SHIPPING_METHOD)
    
    return result
```
